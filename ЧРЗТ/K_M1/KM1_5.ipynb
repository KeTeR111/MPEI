```python
import numpy as np
# Зададим количество аргументов в функции
k = 6
# Левая граница
a = 0
# Правая граница  
b = 1

def func(args: np.array):
    return np.sin(np.sum(args**2))

def func_zad(args):
    ch = args[:3]
    zn = args[3:]
    res = np.sin(np.sum(ch)) / (1 + np.sum(zn**2))
    return res

def integral(f, b, a, k, eps, n_max):
    '''
    Алгоритм работы(Нейронка  просто по русски описала то, что я написал)

    
    Инициализация:

        Задается начальный размер блока точек (n=100)

        Вычисляется объем области интегрирования proiz = (b-a)ᵏ

    Итерационный процесс:

        Генерация: Создается блок случайных точек в k-мерном пространстве

        Вычисление: Для каждой точки считается значение функции

        Накопление: Суммируются значения функции и их квадраты

    Оценка: Вычисляется текущее значение интеграла и погрешность

    Критерии остановки:

        Погрешность становится меньше eps (точность)

        Достигнуто максимальное количество точек n_max

    Адаптация: Размер блока удваивается на каждой итерации для оптимизации производительност
    ''' 
    # Зададим начальное значение n 
    n = 100

    # Зададим произведение перед суммой в формуле интеграла 
    proiz = np.prod(b - a)
    
    # Зададим начальную погрешность 
    err = float('inf')

    # Зададим суммы для подсчета среднего значения функции и среднего значения квадратов функции
    sum = 0
    sum2 = 0
    # Зададим итоговое расчетное число точек 
    total = 0

    # Пока погрешность меньше заданной и total < n max
    while err > eps and total < n_max:
        
        block = min(n, n_max - total)
        total += block

        # В цикле считаем функцию от xi и накапливаем сумму для подсчета среднего значения функции и среднего значения квадратов функции
        for _ in range(block):
            xi_n = np.random.uniform(a, b, k)
            fu = f(xi_n)
            sum += fu
            sum2 += fu**2
        
        # Считаем среднее значений результатов функции и результатов квадратов функции при total(количество) точках
        func_mean =  sum / total
        func2_mean = sum2 / total

        # Считаем интеграл по формуле 
        I = proiz / total * sum

        # Считаем погрешность по формуле 
        err = 1.96 * ((func2_mean - func_mean**2) / total)**0.5
        
        # Увеличивааем количество прибавляемых точех к total и ставим ограничение на максимальный блок как половину n_max
        n = min(n * 2, n_max // 2)

    return I, err, total

I, err, n = integral(f=func_zad, b=b, a=a, k=k, eps=0.00001, n_max=10**6)
I, err, n
        
```
